```python
lm = smf.ols("Q('Annual Fuel1 Cost - Conventional Fuel') ~ Q('City FE (Guide) - Conventional Fuel') + Q('# Cyl')", data = cars).fit()
lm.summary()
```

```python
sns.regplot(y="Annual Fuel1 Cost - Conventional Fuel", x="City FE (Guide) - Conventional Fuel", data=cars, fit_reg = True)
```

```python
lm.resid
```

```python
lm.resid.hist(bins= 15)
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.title("Residuals of linear model")
```

```python
lm.fittedvalues
```

```python
plt.scatter(cars['Annual Fuel1 Cost - Conventional Fuel'], lm.fittedvalues)
plt.xlabel("Fuel cost")
plt.ylabel("Predicted Fuel cost")
plt.title("Annual Fuel Cost vs Predicted Fuel Cost")
```

The most expensive fuel costs are predicted incorrectly. This is known as a ceiling effect, where the independent variable no longer has an effect on the dependent variable.

```python
plt.scatter(x = cars["Annual Fuel1 Cost - Conventional Fuel"], y = lm.resid)
plt.axhline()
plt.title("Residuals of different fuel cost")
plt.xlabel("Fuel Cost")
plt.ylabel("Residual")
```
