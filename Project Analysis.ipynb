```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
cars = pd.read_csv("e_9-5-2019-no-sales-_9_4_2019public.csv")

cars.head(3)
```

```python
c = ['Division','Mfr Name', 'Model Year', 'Carline', 'Verify Mfr Cd', '# Cyl', 'Transmission', 'City FE (Guide) - Conventional Fuel', 'Hwy FE (Guide) - Conventional Fuel', 'Comb FE (Guide) - Conventional Fuel', 'Oil Viscosity', 'Drive Desc']
cars = pd.read_csv("e_9-5-2019-no-sales-_9_4_2019public.csv", usecols = c)

cars.head(10)
```

```python
cars.info()
```

```python
cars.describe(include = ["O"])
```

```python
maker_counts = cars['Mfr Name'].value_counts()
maker_counts
```

```python
maker_counts.plot.bar()
plt.title("Car Makers Distribution")
plt.xlabel("car manufacturers")
plt.ylabel("# of Cars")
```

```python
cars['City FE (Guide) - Conventional Fuel'].hist(bins = 25)
plt.title("City Fuel Milage")
plt.xlabel("miles per gallon")
plt.ylabel("# of cars")
```

```python
cars['Hwy FE (Guide) - Conventional Fuel'].hist(bins = 25)
plt.title("Highway Fuel Milage")
plt.xlabel("miles per gallon")
plt.ylabel("# of cars")
```

```python
cars['Comb FE (Guide) - Conventional Fuel'].hist(bins = 25)
plt.title("Combination Fuel Milage")
plt.xlabel("miles per gallon")
plt.ylabel("# of cars")
```

```python
cars['# Cyl'].hist(bins = 15)
plt.title("Number of Cylinder")
plt.xlabel("# of cylinders")
plt.ylabel("# of cars")
```

```python
trans_counts = cars['Transmission'].value_counts()
trans_counts
```

```python
trans_counts.plot.bar()
plt.title("Transmission Distribution")
plt.xlabel("Type of transmission")
plt.ylabel("# of Cars")
```

```python
sns.relplot(y ="Hwy FE (Guide) - Conventional Fuel", x = "City FE (Guide) - Conventional Fuel", hue = "Mfr Name", data = cars)
plt.title("City vs Highway Fuel Efficiency")
```

```python
sns.relplot(x ="# Cyl", y = "City FE (Guide) - Conventional Fuel", hue = "Mfr Name", data = cars)
plt.title("# of Cylinders relationship to fuel efficiency")
```

```python
g =sns.relplot(x ="Drive Desc", y = "City FE (Guide) - Conventional Fuel", hue = "Mfr Name", data = cars,size=2, aspect=2)
plt.title("Drive System relation to fuel efficiency")
```
